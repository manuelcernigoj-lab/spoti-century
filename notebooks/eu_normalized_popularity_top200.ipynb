#### 📌 We aim to calculate the popularity of tracks in EU by using the streams of top200 chart tracks as a proxy (representing the 78% of total tracks).
#### Viral 50 charts represent the 22% of the total tracks in the dataset and we choose to ignore it for this analysis, since it doesn't have streams data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# file created in 02_charts_elt.ipynb
df_eu_top200 = pd.read_csv('../data/eu_top200_with_tracks_data.csv')
```

##### ⚠️ The popularity column in the tracks datasets is a global score (1-100), not region based, so if we want an Eu-focused analysis, we need to calculate the percentage of `streams` (of the chartes.csv dataset) for each track in Europe (how many times a track has been streamed compared to the total streams)

```python
total_streams = df_eu_top200['streams'].sum()

streams_by_track = df_eu_top200.groupby('track_id')['streams'].sum() / total_streams
# Scale to 0-100 using min-max normalization
normalized_streams = (streams_by_track - streams_by_track.min()) / (streams_by_track.max() - streams_by_track.min()) * 100
normalized_streams.name = "normalized_popularity"
normalized_streams.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(normalized_streams, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Popularity (Min-Max Scaled 0-100)')
plt.xlabel('Popularity Score (0-100)')
plt.yscale('log')
plt.ylabel('Number of Tracks')
plt.tight_layout()
plt.show()
```

```python
df_eu_top_200_with_popularity = df_eu_top200.merge(normalized_streams, on="track_id")
df_eu_top_200_with_popularity.sort_values(by="normalized_popularity", ascending=False).head(20)
```

```python
df_eu_top_200_with_popularity.to_csv("../data/eu_top200_with_custom_popularity.csv")
```
